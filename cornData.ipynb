# Data Cleaning

```python
# drop unnecessary column
data=data.drop('County', axis=1)
```

```python
# dealing with missing Education level
data['Education']= data['Education'].replace(np.nan, 'No education')
```

## Predict Missing Acreage

```python
# check missing acreage
data.Acreage.isnull().sum()
```

```python
# create numeric columns for categorical columns
data['education_code']=data['Education'].astype('category').cat.codes
data['Gender_code']=data['Gender'].astype('category').cat.codes
data['age_bracket_code']=data['Age bracket'].astype('category').cat.codes
data['main_credit_source_code']=data['Main credit source'].astype('category').cat.codes
data['farm_records_code']=data['Farm records'].astype('category').cat.codes
data['main_advisory_source_code']=data['Main advisory source'].astype('category').cat.codes
data['extension_provider_code']=data['Extension provider'].astype('category').cat.codes
data['advisory_format_code']=data['Advisory format'].astype('category').cat.codes
data['advisory_language_code']=data['Advisory language'].astype('category').cat.codes
```

```python
from sklearn.linear_model import LinearRegression

# Assuming 'x' contains features used for prediction and 'y' contains target 'Acreage'
x = data.dropna(subset=['Acreage']).drop(['Acreage','Farmer','Crop','Power source','Water source','Crop insurance', 'Education', 'Gender', 'Age bracket','Main credit source','Farm records', 'Main advisory source', 'Extension provider',
       'Advisory format', 'Advisory language'], axis=1)
y = data.dropna(subset=['Acreage'])['Acreage']
```

```python
# join x and y
xy=x.join(y)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# plot a heatmap for correlation
plt.figure(figsize=(15,6))
sns.heatmap(xy.corr(),cmap='YlGnBu')
```

```python
# Fit a regression model
model = LinearRegression()
model.fit(x, y)
```

```python
# Predict missing values
missing_x = data[data['Acreage'].isnull()].drop(['Acreage','Farmer','Crop','Power source','Water source','Crop insurance', 'Education', 'Gender', 'Age bracket','Main credit source','Farm records', 'Main advisory source', 'Extension provider',
       'Advisory format', 'Advisory language'], axis=1)
predicted_acreage = model.predict(missing_x)
```

```python
# Replace missing values with predicted values
data.loc[data['Acreage'].isnull(), 'Acreage'] = predicted_acreage
```

```python
# find outliers for Fertilizer amount
Q1= np.percentile(data['Fertilizer amount'], 25)
Q3= np.percentile(data['Fertilizer amount'], 75)

lower_bound= Q1- 1.5*(Q3-Q1)
upper_bound= Q3+ 1.5*(Q3-Q1)
upper_bound
```

```python
# cap fertilizer amount at 300
mask= data['Fertilizer amount']>300
data.loc[mask, 'Fertilizer amount']=300
```
